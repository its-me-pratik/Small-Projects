```python
#First we run a RANDOMISED SEARCH to find relevant parameters to tune
from sklearn.model_selection import RandomizedSearchCV
#No. of trees to consider in the forest
n_estimators = [int(x) for x in np.linspace(start = 200, stop = 2000, num = 10)]
#No. of features to consider at every split
max_features = ['auto', 'sqrt']
#Max no. of levels in a tree
max_depth = [int(x) for x in np.linspace(10, 110, num = 11)]
max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 4]
# Method of selecting samples for training each tree
bootstrap = [True, False]
```

We use __Randomised search__ here because grid search would be too slow because of all the different models to train.
__Grid search__ would be too slow so we first run a randomised search and then furthur run a grid search.

```python
#Creating the random grid
random_grid = {'n_estimators': n_estimators, 'max_features': max_features, 'max_depth': max_depth, 'min_samples_split': min_samples_split, 'min_samples_leaf': min_samples_leaf, 'bootstrap': bootstrap}
pprint(random_grid)
```

```python
#Creating base model to tune
rf = RandomForestRegressor()
rf_random = RandomizedSearchCV(estimator = rf, param_distributions = random_grid, n_iter = 60, cv = 3, verbose = 2, random_state = 42, n_jobs = -1)
#Fit random search model
rf_random.fit(X_train, Y_train)
rf_random.best_params_
```

```python
#Evaluating Random Search Performance
best_random = rf_random.best_estimator_
random_accuracy = evaluate(best_random, X_train, Y_train)
print(best_random)
```

```python
#Getting results using k-fold Cross Validation
from sklearn.model_selection import cross_val_score
accuracies_RFR_tuned = cross_val_score(estimator = best_random,X = X_train,y = Y_train, cv = 10)
Accuracy_RFR_tuned = accuracies_RFR_tuned.mean()
accuracies_RFR_tuned.std()
print(Accuracy_RFR_tuned)
```
